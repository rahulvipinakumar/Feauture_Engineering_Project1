```python
import sys; 
sys.path.insert(0, '..')

import findspark
findspark.init()
from pyspark.sql import SparkSession
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#import nltk
#nltk.download()

%matplotlib inline

spark = SparkSession.builder.appName("pyspark-2").getOrCreate()

df_spark = spark.read.options(inferSchema='True').option('escape','"').csv("/dataset/nyc-jobs.csv", header=True)
#Converting to Pandas for easy analysis
df=df_spark.toPandas()
#df_spark.printSchema()
```

```python
#df_spark.groupby("Posting Type").count().orderBy("count", ascending=False).first()[0]
from pyspark.sql.window import Window
import pyspark.sql.functions as f
#df_spark_tmp=df_spark.groupby(["Posting Type","Agency","Business Title","Full-Time/Part-Time indicator"]).count().orderBy("count", ascending=False)
#windowSpec  = Window.partitionBy(["Posting Type","Agency","Business Title"]).orderBy(f.col("count").desc())

def gap_filling_by_mode(df_spark,grp_by_cols,req_col):
    grp_by_cols_copy=grp_by_cols.copy()
    grp_by_cols_copy.append(req_col)
    df_spark_tmp=df_spark.groupby(grp_by_cols_copy).count().orderBy("count", ascending=False)
    windowSpec  = Window.partitionBy(grp_by_cols).orderBy(f.col("count").desc())
    df_spark_tmp_final=df_spark_tmp.withColumn("rank",f.row_number().over(windowSpec))
    df_spark_tmp_final=df_spark_tmp_final.filter(f.col("rank") == 1).withColumnRenamed(req_col,req_col+"_new")
    #abc=df_spark_tmp_final.toPandas()
    #abc.isnull().sum()
    main_cols=df_spark.columns
    df_spark=df_spark.join(df_spark_tmp_final, on=grp_by_cols, how='left')
    df_spark=df_spark.withColumn(req_col,f.coalesce(f.col(req_col),f.col(req_col+"_new")))
    return df_spark[[main_cols]]

df_spark=gap_filling_by_mode(df_spark,['Salary Frequency'],'Full-Time/Part-Time indicator')

df_spark=gap_filling_by_mode(df_spark,['Posting Type','Agency'],'Posting Date')
```

### Creating additional columns
#### Challenges:
##### 1) Salary Frequency is not uniform, so its better to bring it to a common frequency. Here I took, DAILY as common frequency.
##### 2) Salary can be used for analysis purpose if its a fixed values instead of range.
##### 3) Year and month can be extracted from POSTING DATE and PROCESS DATE for analysis in year and month level
##### 4) Fill the missing values 

### Additional columns for salary, salary_avg, daily_salary_avg

```python
from pyspark.sql.types import IntegerType,BooleanType,DateType
from pyspark.sql.functions import when
# Will convert the salary columns to Integer type for performing aggregate functions
df_spark = df_spark.withColumn("Salary Range To",df_spark["Salary Range To"].cast(IntegerType()))
df_spark = df_spark.withColumn("Salary Range From",df_spark["Salary Range From"].cast(IntegerType()))
# Finding average salary from from and To salaries
df_spark = df_spark.withColumn("salary_avg",(df_spark["Salary Range From"]+df_spark["Salary Range To"])/2.0)
#Normalizing the salary to a frequency of Daily
df_spark = df_spark.withColumn("daily_salary_avg", when(df_spark["Salary Frequency"] == "Annual",df_spark["salary_avg"]/(12*22.00))\
                             .when(df_spark["Salary Frequency"] == "Hourly",df_spark["salary_avg"]*8.00)\
                             .otherwise(df_spark["salary_avg"]))
```

### Additional columns for month and year

```python
# column with Year and month details of Posting Date
df_spark=df_spark.withColumn("Job_Post_Year",f.year(f.to_timestamp(df_spark["Posting Date"],'yyyy-MM-dd hh24:mi:ss.ff3'))\
                               .cast(IntegerType())).withColumn("Job_Post_Month",\
                                                                f.month(f.to_timestamp(df_spark["Posting Date"],'yyyy-MM-dd hh24:mi:ss.ff3')))

# column with Year and month details of Process Date
df_spark=df_spark.withColumn("data_processed_year",f.year(f.to_timestamp(df_spark["Process Date"],'yyyy-MM-dd hh24:mi:ss.ff3'))\
                               .cast(IntegerType())).withColumn("data_processed_month",\
                                                                f.month(f.to_timestamp(df_spark["Process Date"],'yyyy-MM-dd hh24:mi:ss.ff3')))
```

What's the number of jobs posting per category (Top 10)?

```python
import pyspark.sql.functions as f
from pyspark.sql.types import IntegerType,BooleanType,DateType

def sum_per_category(dataframe,category_col,aggregate_col,alias_col_name,num_of_output_recs):
    # dataframe : The input dataframe
    # category_col : The column name used for categorizing the aggregate functionality
    # aggregate_col : The column name used for performing Aggregate functions
    # alias_col_name : The alias column name for the aggragated results
    # num_of_output_recs : Number of result records
    
    # Will convert the aggregate column to Integer type for performing aggregate functions
    df_final = dataframe.withColumn(aggregate_col,df[aggregate_col].cast(IntegerType()))
    
    #Returning final result with aggregated data
    return df_final.groupBy(category_col).agg(f.sum(aggregate_col).alias(alias_col_name)).sort(f.col(alias_col_name).desc()).limit(num_of_output_recs)

df_Num_job_posting = sum_per_category(df,"Job Category","# Of Positions","Num_Job_Posting",10)
df_Num_job_posting.show(truncate=False)
```

```python
import pyspark.sql.functions as f
from pyspark.sql.types import IntegerType,BooleanType,DateType

def sal_dist_per_category(dataframe,category_name):
    # dataframe : The input dataframe
    # category_name : The category name for which salary distribution need to be calculated
    
    # Will convert the salary columns to Integer type for performing aggregate functions
    df_temp = dataframe.withColumn("Salary Range To",dataframe["Salary Range To"].cast(IntegerType()))
    df_temp = df_temp.withColumn("Salary Range From",df_temp["Salary Range From"].cast(IntegerType()))
    # Finding average salary from from and To salaries
    df_temp = df_temp.withColumn("salary_avg",(df_temp["Salary Range From"]+df_temp["Salary Range To"])/2.0)
    #Normalizing the salary to a frequency of Daily
    df_temp = df_temp.withColumn("daily_salary_avg", when(df_temp["Salary Frequency"] == "Annual",df_temp["salary_avg"]/(12*22.00))\
                                 .when(df_temp["Salary Frequency"] == "Hourly",df_temp["salary_avg"]*8.00)\
                                 .otherwise(df_temp["salary_avg"]))
    
    # Will filter data based on given category and convert to pandas dataframe
    df_pandas = df_temp.filter(f.col("Job Category") == category_name).toPandas()
    
    #Ploting histogram of salary distribution
    df_pandas.hist(column="daily_salary_avg")

sal_dist_per_category(df,"Information Technology & Telecommunications")
```

What's the job posting having the highest salary per agency?

```python
import pyspark.sql.functions as f
from pyspark.sql.window import Window
from pyspark.sql.functions import when
from pyspark.sql.types import IntegerType,BooleanType,DateType

def max_sal_per_agency(dataframe):
    # dataframe : The input dataframe
    
    # Will convert the salary columns to Integer type for performing aggregate functions
    df_max_final = dataframe.withColumn("Salary Range To",dataframe["Salary Range To"].cast(IntegerType()))
    df_max_final = df_max_final.withColumn("Salary Range From",dataframe["Salary Range From"].cast(IntegerType()))
    # Finding average salary from from and To salaries
    df_max_final = df_max_final.withColumn("Salary_Avg",(df_max_final["Salary Range From"]+df_max_final["Salary Range To"])/2.0)

    #Normalizing the salary to a frequency of Daily
    df_max_final = df_max_final.withColumn("daily_salary_avg", when(df_max_final["Salary Frequency"] == "Annual",\
                                                                    df_max_final["salary_avg"]/(12*22.00))\
                                 .when(df_max_final["Salary Frequency"] == "Hourly",df_max_final["salary_avg"]*8.00)\
                                 .otherwise(df_max_final["salary_avg"]))
    
    
    # Window function definition
    windowSpec  = Window.partitionBy("Agency").orderBy(f.col("Salary_Avg").desc())
    
    #Adding Rank based on avg salary in desc order
    df_max_final=df_max_final.withColumn("rank",f.rank().over(windowSpec))

    #Output the records having max salary for each agency
    return df_max_final.filter(f.col("rank") == 1)

df_max_sal_per_agency = max_sal_per_agency(df)
df_max_sal_per_agency.show()
```

What's the job postings average salary per agency for the last 2 years?

```python
import pyspark.sql.functions as f
from pyspark.sql.functions import when
def avg_sal_per_agency(dataframe):
    # dataframe : The input dataframe
    
    # Will convert the salary columns to Integer type for performing aggregate functions
    df_temp = dataframe.withColumn("Salary Range To",dataframe["Salary Range To"].cast(IntegerType()))
    df_temp = df_temp.withColumn("Salary Range From",df_temp["Salary Range From"].cast(IntegerType()))
    # Finding average salary from from and To salaries
    df_temp = df_temp.withColumn("salary_avg",(df_temp["Salary Range From"]+df_temp["Salary Range To"])/2.0)
    #Normalizing the salary to a frequency of Daily
    df_temp = df_temp.withColumn("daily_salary_avg", when(df_temp["Salary Frequency"] == "Annual",df_temp["salary_avg"]/(12*22.00))\
                                 .when(df_temp["Salary Frequency"] == "Hourly",df_temp["salary_avg"]*8.00)\
                                 .otherwise(df_temp["salary_avg"]))
    
    
    #Finding the Year of job posting aand year of data processed
    df_temp=df_temp.withColumn("Job_Post_Year",f.year(f.to_timestamp(df_temp["Posting Date"],'yyyy-MM-dd hh24:mi:ss.ff3'))\
                               .cast(IntegerType())).withColumn("data_processed_year",\
                                                                f.year(f.to_timestamp(df_temp["Process Date"],'yyyy-MM-dd hh24:mi:ss.ff3'))\
                                                                .cast(IntegerType()))
    
    #FIlter data for last 2 years
    df_temp = df_temp.filter(df_temp["Job_Post_Year"]>=df_temp["data_processed_year"]-1)
    
    #Average slaray per agency in last 2 years
    return df_temp.groupBy("Agency").agg(f.avg("daily_salary_avg").alias("avg_daily_salary_per_agency"))

df_avg_sal_per_agency = avg_sal_per_agency(df)
df_avg_sal_per_agency.show(truncate=False)
```

What are the highest paid skills in the US market?

```python
import pyspark.sql.functions as f
from pyspark.sql.functions import when
from pyspark.sql.types import IntegerType,BooleanType,DateType

def max_sal_per_skill(dataframe,num_of_output_recs):
    # dataframe : The input dataframe
    # num_of_output_recs : Number of result records
    
    # Will convert the salary columns to Integer type for performing aggregate functions
    df_temp = dataframe.withColumn("Salary Range To",dataframe["Salary Range To"].cast(IntegerType()))
    df_temp = df_temp.withColumn("Salary Range From",df_temp["Salary Range From"].cast(IntegerType()))
    # Finding average salary from from and To salaries
    df_temp = df_temp.withColumn("salary_avg",(df_temp["Salary Range From"]+df_temp["Salary Range To"])/2.0)
    #Normalizing the salary to a frequency of Daily
    df_temp = df_temp.withColumn("daily_salary_avg", when(df_temp["Salary Frequency"] == "Annual",df_temp["salary_avg"]/(12*22.00))\
                                 .when(df_temp["Salary Frequency"] == "Hourly",df_temp["salary_avg"]*8.00)\
                                 .otherwise(df_temp["salary_avg"]))
    
    #Returning final result with aggregated data
    return df_temp.groupBy("Job Category").agg(f.max("daily_salary_avg").alias("max_daily_salary")).sort(f.col("max_daily_salary").desc()).limit(num_of_output_recs)

df_max_sal = max_sal_per_skill(df_spark,10)
df_max_sal.show(truncate=True)
```
